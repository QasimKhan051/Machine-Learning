```python
#loading Dataset 
b_cancer=load_breast_cancer()
```

```python
#---Question 0 (Example)

def answer_zero():
    return len(b_cancer['feature_names'])
answer_zero() 
```

```python
##---Question 1
def answer_one():
    return pd.DataFrame(data=np.c_[b_cancer.data, b_cancer.target], columns=list(b_cancer.feature_names) + ['target'])
answer_one() 
```

```python
#---Question 2
def answer_two():
    b_cancer_dataframe = answer_one()
    df = b_cancer_dataframe['target'].value_counts()
    df.index = ['benign', 'malignant']
    return df
answer_two()
```

```python
#---Question 3
def answer_three():
    b_cancer_dataframe = answer_one()
    X = b_cancer_dataframe.iloc[:, 0:30]
    y = b_cancer_dataframe['target']
    return X, y
answer_three()
```

```python
#-----Question 4
from sklearn.model_selection import train_test_split
def answer_four():
    X, y = answer_three()
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    return X_train, X_test, y_train, y_test
answer_four()
```

```python
#----Question 5
from sklearn.neighbors import KNeighborsClassifier

def answer_five():
    X_train, X_test, y_train, y_test = answer_four()
    b_knn = KNeighborsClassifier(n_neighbors = 1)
    b_knn.fit(X_train, y_train)
    return b_knn
answer_five()
```

```python
#----Question 6
def answer_six():
    b_cancer_dataframe = answer_one()
    means = b_cancer_dataframe.mean()[:-1].values.reshape(1, -1)
    b_knn = answer_five()
    return b_knn.predict(means)
answer_six()
```
